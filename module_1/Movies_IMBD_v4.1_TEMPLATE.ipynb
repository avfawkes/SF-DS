# 24. Какая студия дает самые длинные названия своим фильмам по количеству символов?

```python
data_f = data.copy()
data_f.production_companies = data_f.production_companies.str.split('|')

data_f2 = data_f.explode('production_companies').groupby('production_companies'). \
agg({'original_title':['count', lambda x:len(''.join(x))]})

data_f2.columns = ['count', 'str_len']
data_f2['mean_len'] = data_f2['str_len']/data_f2['count']
data_f2.sort_values('mean_len', ascending=False)
```

ВАРИАНТ 2

```python
start = time.time()

film_len_df = pd.DataFrame(columns = ['title_cum_len','film_count'])
for row in data.itertuples():
    for studio in row.production_companies.split(sep="|"):
        if studio in film_len_df.index:
            film_len_df.title_cum_len.loc[studio] += len(row.original_title)
            film_len_df.film_count.loc[studio] += 1
        else:
            film_len_df = film_len_df.append \
            (pd.Series({'title_cum_len':len(row.original_title), 'film_count':1}, name=studio))

film_len_df['av_len'] = film_len_df['title_cum_len']/film_len_df['film_count']
display(film_len_df.sort_values('av_len', ascending=False))

stop = time.time()
print('Время выполнения:', round(stop-start, 2), 'сек.')
```

ВАРИАНТ 3

```python
start = time.time()

studio_set = set()
data.production_companies.apply(lambda item: studio_set.update(item.split(sep="|")))

studio_str_cum_len = dict.fromkeys(studio_set, 0)
for studio in studio_set:
    studio_str_cum_len[studio] = len(''.join(data[data.production_companies.str.contains(studio)].original_title))
    
studio_film_count = dict.fromkeys(studio_set, 0)
for studio in studio_set:
    studio_film_count[studio] = data[data.production_companies.str.contains(studio)].original_title.count()

studio_res = {key: studio_str_cum_len[key]/studio_film_count.get(key, 1) for key in studio_str_cum_len.keys()}

print(max(studio_res, key=studio_res.get))

stop = time.time()
print('Время выполнения:', round(stop-start, 2), 'сек.')
```

# 25. Описание фильмов какой студии в среднем самые длинные по количеству слов?

```python
data_f3 = data_f.explode('production_companies').groupby('production_companies'). \
agg({'overview':['count', lambda x:len(' '.join(x).split())]})
data_f3.columns = ['count', 'words_count']
data_f3['mean_words'] = data_f3['words_count']/data_f3['count']
data_f3.sort_values('mean_words', ascending=False)
```

ВАРИАНТ 2

```python
start = time.time()

film_len_df = pd.DataFrame(columns = ['title_cum_len','overview_cum_word','film_count'])
for row in data.itertuples():
    for studio in row.production_companies.split(sep="|"):
        if studio in film_len_df.index:
            film_len_df.title_cum_len.loc[studio] += len(row.original_title)
            film_len_df.overview_cum_word.loc[studio] += len(row.overview.split())
            film_len_df.film_count.loc[studio] += 1
        else:
            film_len_df = film_len_df.append \
            (pd.Series({'title_cum_len':len(row.original_title),
                        'overview_cum_word':len(row.overview.split()),
                        'film_count':1}, name=studio))

film_len_df['av_len'] = film_len_df['title_cum_len']/film_len_df['film_count']
film_len_df['av_word'] = film_len_df['overview_cum_word']/film_len_df['film_count']
display(film_len_df.sort_values('av_word', ascending=False))

stop = time.time()
print('Время выполнения:', round(stop-start, 2), 'сек.')
```
